```python
import pandas as pd
import pickle
```

```python
dataset_till_2020_csv = "population_data_till_2020.csv"
dataset_2021_2022_csv = "population_data_2021_2022.csv"
dataset_pickle = "population_data.pickle"
```

```python
data_till_2020: pd.DataFrame = pd.read_csv(dataset_till_2020_csv)
```

```python
def condense_age(age):
    if age >= 65:
        return '65 years or older'
    elif 50 <= age <= 64:
        return '50-64 years'
    else:
        return 'Overall'
```

```python
data_till_2020['AGE_GROUP'] = data_till_2020['AGE'].apply(condense_age)
```

```python
data_till_2020["RACE"] = data_till_2020.apply(lambda row: -1 if row['ORIGIN'] == 2 else row["RACE"], axis=1)
data_till_2020 = data_till_2020.drop(columns=['ORIGIN'])
```

```python
to_drop = ["DIVISION", "REGION", "SUMLEV", "STATE", "CENSUS2010POP", "ESTIMATESBASE2010", "POPESTIMATE2012",
    "POPESTIMATE2013", "POPESTIMATE2014", "POPESTIMATE2010", "POPESTIMATE2011", "AGE"]
data_till_2020 = data_till_2020.drop(columns=to_drop)
```

```python
data_till_2020 = data_till_2020.groupby(["AGE_GROUP", "NAME", "SEX", "RACE"]).agg("sum").reset_index()
```

```python
data_2020_2021: pd.DataFrame = pd.read_csv(dataset_2021_2022_csv)
```

```python
data_2020_2021['AGE_GROUP'] = data_2020_2021['AGE'].apply(condense_age)
```

```python
data_2020_2021["RACE"] = data_2020_2021.apply(lambda row: -1 if row['ORIGIN'] == 2 else row["RACE"], axis=1)
data_2020_2021 = data_2020_2021.drop(columns=['ORIGIN'])
```

```python
to_drop = ["DIVISION", "REGION", "SUMLEV", "STATE", "POPESTIMATE2020", "ESTIMATESBASE2020", "ESTIMATESBASE2020", "AGE"]
data_2020_2021 = data_2020_2021.drop(columns=to_drop)
```

```python
data_2020_2021 = data_2020_2021.groupby(["AGE_GROUP", "NAME", "SEX", "RACE"]).agg("sum").reset_index()
```

```python
data_2020_2021
```

```python
data_all = data_till_2020
data_all[["POPESTIMATE2021", "POPESTIMATE2022"]] = data_2020_2021[["POPESTIMATE2021", "POPESTIMATE2022"]]
```

```python
pickle.dump(data_all, open(dataset_pickle, "wb"))
```
